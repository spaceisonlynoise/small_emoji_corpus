```python
#make it lists of strings
emotional_posts_emojis = []
for post in new_list:
    emotional_posts_emojis.append([post])
print(len(emotional_posts_emojis)) #730
```

```python
import random
#get 730 posts without any emoji
rand_posts_without_emojis = []
for i in range(635):
    rand_posts_without_emojis.append(posts_without_emojis_dupl.pop(random.randrange(len(posts_without_emojis_dupl))))
print(len(rand_posts_without_emojis))
```

```python
#construct tupels
tupls_posts = []
for list_ in emotional_posts_emojis:
    for string_ in list_:
        tupls_posts.append((string_, 1))
for post in rand_posts_without_emojis:
    tupls_posts.append((post, 0))
tupls_posts
```

```python
#check tuples for empty post because of division by zero bug in wordvector
for tup in tupls_posts:
    if len(tup[0]) == 1:
        tupls_posts.remove(tup)
    #print(tup[0], len(tup[0]))
```

```python
#now lists with posts and labels
emot_texts = []
emot_labels = []
for tupl in tupls_posts:
    emot_texts.append(tupl[0])
    emot_labels.append(tupl[1])
emot_texts = [''.join(x) for x in emot_texts]
```

```python
for text in emot_texts:
    if len(text) == 0:
        print("yes")
```

```python
#construct clean version to train with
#remove all emojis
clean_texts = []
clean_post = ""
for post in emot_texts:
    for c in post:
        if c not in emoji.UNICODE_EMOJI:
            clean_post += c
    clean_texts.append(" ".join(clean_post.split())) #remove redundant white space
    clean_post = ""


emot_texts = clean_texts
emot_texts
```
